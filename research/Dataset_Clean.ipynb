```python
import pandas as pd
```

```python


# Read dataset
df = pd.read_csv(r"D:\How_Educated_India\research\Marged_Dataset.csv")

df
```

```python
# Drop rows where 'Area Name' is 'India'
df_filtered = df[df['State Name'] != "INDIA"]

# Reset index
df_filtered = df_filtered.reset_index(drop=True)

df_filtered
```

```python


# Count unique states per year
unique_states_per_year = df_filtered.groupby("Year")["State Name"].nunique().reset_index()

# Rename columns for clarity
unique_states_per_year.columns = ["Year", "Unique_States"]

print(unique_states_per_year)
```

```python


# Assuming your DataFrame has 'Year' and 'State Name' columns
# Filter states for each year
states_1991 = set(df_filtered[df_filtered["Year"] == 1991]["State Name"])
states_2001 = set(df_filtered[df_filtered["Year"] == 2001]["State Name"])

# Find states in 2001 but not in 1991
new_states_2001 = states_2001 - states_1991

print("States present in 2001 but not in 1991:")
print(new_states_2001)
```

```python
# Filter states for each year
states_1991 = set(df_filtered[df_filtered["Year"] == 1991]["State Name"])
states_2011 = set(df_filtered[df_filtered["Year"] == 2011]["State Name"])

# Find states in 2011 but not in 1991
new_states_2011 = states_2011 - states_1991

print("States present in 2011 but not in 1991:")
print(new_states_2011)
```

```python


# Group by state name and count rows
state_counts = df_filtered.groupby("State Name").size().reset_index(name="Count")

# Sort by count if you want
state_counts = state_counts.sort_values(by="Count", ascending=False)

print(state_counts)
```

```python
state_counts = df_filtered.groupby("State Code").size().reset_index(name="Count")

print(state_counts)
```

```python
# Group by Year and get unique states as a list
unique_states_per_year = df_filtered.groupby("Year")["State Name"].unique().reset_index()

# Rename column for clarity
unique_states_per_year.columns = ["Year", "Unique_States"]

print(unique_states_per_year)
```

```python

# Group by Year and get sets of unique states per year
states_per_year = df_filtered.groupby("Year")["State Code"].apply(set)

# Get intersection of all years (states present in every year)
states_in_all_years = set.intersection(*states_per_year)

# Get all states that appear at least once
all_states = set.union(*states_per_year)

# States NOT present in every year
states_not_in_all_years = all_states - states_in_all_years

print("States not present in every year:")
states_not_in_all_years
```

```python
# Remove rows where State Code is 13, 31, or 35
df_filtered = df_filtered[~df_filtered["State Code"].isin([13, 31, 35])]

# Check the result
df_filtered
```

```python
# Count unique states per year
unique_states_per_year = df_filtered.groupby("Year")["State Name"].nunique().reset_index()

# Rename columns for clarity
unique_states_per_year.columns = ["Year", "Unique_States"]

print(unique_states_per_year)
```

```python
df_filtered ["State Name"] = df_filtered ["State Name"].str.replace(r"^State\s*-\s*|\s*\d+$", "", regex=True)

# Check the result
df_filtered ["State Name"].head()
```

```python
# Group by Year and get sets of unique states per year
states_per_year = df_filtered.groupby("Year")["State Name"].apply(set)

# Get intersection of all years (states present in every year)
states_in_all_years = set.intersection(*states_per_year)

# Get all states that appear at least once
all_states = set.union(*states_per_year)

# States NOT present in every year
states_not_in_all_years = all_states - states_in_all_years

print("States not present in every year:")
states_not_in_all_years
```

```python


# Define a mapping of old state names to new standardized names
state_replacements = {
    "NCT OF DELHI": "DELHI",
    "ORISSA": "ODISHA",
    "PUDUCHERRY": "PONDICHERRY",
    "UTTARANCHAL": "UTTARAKHAND"
}

# Replace state names in the 'State Name' column
df_filtered["State Name"] =  df_filtered["State Name"].replace(state_replacements)

# Check the result
print( df_filtered["State Name"].unique())
```

```python
# Group by Year and get sets of unique states per year
states_per_year = df_filtered.groupby("Year")["State Name"].apply(set)

# Get intersection of all years (states present in every year)
states_in_all_years = set.intersection(*states_per_year)

# Get all states that appear at least once
all_states = set.union(*states_per_year)

# States NOT present in every year
states_not_in_all_years = all_states - states_in_all_years

print("States not present in every year:")
states_not_in_all_years
```
