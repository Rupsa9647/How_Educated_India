```python
import pandas as pd
```

```python


# Define your columns as a list
columns = [
    "Table Name", "State Code", "Distt Code", "Tehsil Code", "State Name", "Area", "Age",
    "Total Person", "Total Males", "Total Females",
    "Illiterate Person", "Illiterate Males", "Illiterate Females",
    "Literate Person", "Literate Males", "Literate Females",
    "Literate without educational level Person", "Literate without educational level Males", "Literate without educational level Females",
    "Below primary Person", "Below primary Males", "Below primary Females",
    "Primary Person", "Primary Males", "Primary Females",
    "Middle Person", "Middle Males", "Middle Females",
    "Secondary Person", "Secondary Males", "Secondary Females",
    "Higher secondary Person", "Higher Secondary Males", "Higher Secondary Females",
    "Non-technical diploma Person", "Non-technical diploma Males", "Non-technical diploma Females",
    "Technical diploma Person", "Technical diploma Males", "Technical diploma Females",
    "Graduate Person", "Graduate Males", "Graduate Females",
    "Unclassified Person", "Unclassified Males", "Unclassified Females"
]

# Load the CSV
df_2001 = pd.read_excel(r"D:\How_Educated_India\Datasets\2001.xls", 
                 skiprows=6,    # Skip first 6 rows
                 names=columns, # Assign column names
                 index_col=False) # Keep default index (0,1,2,...)
df_2001["Year"]=2001
df_2001 = df_2001.drop("Tehsil Code", axis=1)

df_2001.to_csv("clean_dataset_2001.csv")
df_2001
```

```python


df_1991 = pd.read_excel(r"D:\How_Educated_India\Datasets\1991.xls", 
                 skiprows=6,    # Skip first 6 rows
                 names=columns, # Assign column names
                 index_col=False) # Keep default index (0,1,2,...)

df_1991["Year"]=1991
df_1991 = df_1991.drop("Tehsil Code", axis=1)
df_1991.to_csv("clean_dataset_1991.csv")
df_1991
```
